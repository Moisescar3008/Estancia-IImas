Logistic Regression

```python
import pandas as pd
train = pd.read_csv('Train_100.csv')
train
```

```python
val = pd.read_csv('Val_100.csv')
val
```

```python
test = pd.read_csv('Test_100.csv')
test
```

```python
X_train = train[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_train = train["etiquetas"]
```

```python
X_val = val[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_val = val["etiquetas"]
```

```python
X_test = test[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_test = test["etiquetas"]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
from sklearn.linear_model import LogisticRegression

param_grid = {
    'C': [0.1, 1, 10],
    'penalty': ['l2'],
    'multi_class': ['multinomial', 'ovr'],
    'solver': ['newton-cg', 'sag' ],
    'max_iter': [100, 500, 1000]
}
```

```python
lr_model = LogisticRegression()
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=False)
```

```python
grid_search = GridSearchCV(estimator=lr_model, param_grid=param_grid, scoring='accuracy', refit=True, cv=cv, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
y_pred_train = grid_search.predict(X_train)
```

```python
train_lr_accuracy = accuracy_score(y_train, y_pred_train)

print("Logistic Regression Accuracy train:", train_lr_accuracy)
```

```python
from sklearn.metrics import classification_report

train_report = classification_report(y_train, y_pred_train)
print('train report: ', train_report)
```

```python
y_pred_val = grid_search.predict(X_val)
```

```python
val_lr_accuracy = accuracy_score(y_val, y_pred_val)

print("Logistic Regression Accuracy_val:", val_lr_accuracy)
```

```python
val_report = classification_report(y_val, y_pred_val)
print('val report:', val_report)
```

```python
y_pred_test = grid_search.predict(X_test)
```

```python
test_lr_accuracy = accuracy_score(y_val, y_pred_test)

print("Logistic Regression Accuracy_test:", test_lr_accuracy)
```

```python
test_report = classification_report(y_test, y_pred_test)
print('test report:', val_report)
```
