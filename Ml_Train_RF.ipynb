Random Forest

```python
import pandas as pd
train = pd.read_csv('Train_100.csv')
train
```

```python
val = pd.read_csv('Val_100.csv')
val
```

```python
test = pd.read_csv('Test_100.csv')
test
```

```python
X_train = train[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_train = train["etiquetas"]
```

```python
X_val = val[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_val = val["etiquetas"]
```

```python
X_test = test[["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy"]]
y_test = test["etiquetas"]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
param_grid = { 
    'n_estimators': [100],
    'criterion' :['gini', 'entropy'],
    "class_weight":['balanced'],
    "min_impurity_decrease":[1e-4, 1e-7],
    'random_state':[42]
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=False)
```

```python
random_forest = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=cv, refit=True, n_jobs=-1, verbose=2)
```

```python
random_forest.fit(X_train, y_train)
```

```python
random_forest.best_params_
```

```python
y_pred_train = random_forest.predict(X_train) 
```

```python
train_rf_accuracy = accuracy_score(y_train, y_pred_train)

print("accuracy of train in random forest: ", train_rf_accuracy)
```

```python
from sklearn.metrics import classification_report

train_report = classification_report(y_train, y_pred_train)
print('train report: ', train_report)
```

```python
y_pred_val = random_forest.predict(X_val)
```

```python
val_rf_accuracy = accuracy_score(y_val, y_pred_val)

print("accuracy of val in random forest: ", val_rf_accuracy)
```

```python
val_report = classification_report(y_val, y_pred_val)
print('val report: ', val_report)
```
