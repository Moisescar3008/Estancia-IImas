```python
#importar librerias 
import skimage.io as skio
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
from PIL import Image
from skimage.color import rgb2gray
import pandas as pd
import tifffile
```

```python
def imagen2vector_test(imagen1):
    imagen_array = np.array(imagen1)
    v1 = imagen_array.flatten()
    img_vec = v1[:, np.newaxis]
    return img_vec
```

```python
import os
import numpy as np
from skimage import transform
import tifffile

def extract_rgb_features_from_folders(folder_paths, target_shape=(100, 100)):
    features_red = []
    features_green = []
    features_blue = []
    labels = []
    
    classes = ["Sanos", "Agudos", "Cronicos"]
    
    for folder_path in folder_paths:
        for class_name in classes:
            class_path = os.path.join(folder_path, class_name)
            class_index = classes.index(class_name)
            for filename in os.listdir(class_path):
                file_path = os.path.join(class_path, filename)
                image = tifffile.imread(file_path)
                resized_image = transform.resize(image, target_shape)
                    
                red = resized_image[:, :, 0]
                green = resized_image[:, :, 1]
                blue = resized_image[:, :, 2]
                    
                features_red.append(red.flatten())
                features_green.append(green.flatten())
                features_blue.append(blue.flatten())
                labels.append(class_index)  
    features_red = np.array(features_red)
    features_green = np.array(features_green)
    features_blue = np.array(features_blue)
    labels = np.array(labels)
    
    # Repetir las etiquetas para que tengan la misma longitud que las características
    repeated_labels = np.repeat(labels, features_red.shape[1])
    im_label = np.reshape(repeated_labels, (-1, 1))
    
    img_red = imagen2vector_test(features_red)
    img_green = imagen2vector_test(features_green)
    img_blue = imagen2vector_test(features_blue)
    return img_red, img_green, img_blue, im_label

folder_paths = [
    "C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"
]

img_red, img_green, img_blue, im_label = extract_rgb_features_from_folders(folder_paths)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, transform


def extract_hsv_features_from_folders(folder_paths, target_shape=(100, 100)):
    im_hue = []
    im_sat = []
    im_values = []
    classes = ["Sanos", "Agudos", "Cronicos"]
    labels = []
    
    for folder_path in folder_paths:
        for class_name in classes:
            class_path = os.path.join(folder_path, class_name)
            class_index = classes.index(class_name)
            for filename in os.listdir(class_path):
                file_path = os.path.join(class_path, filename)
                image = tifffile.imread(file_path)
                resized_image = transform.resize(image, target_shape)
    
                    # HSV conversion
                hsv_image = color.rgb2hsv(resized_image)
                hue = hsv_image[:, :, 0]
                sat = hsv_image[:, :, 1]
                values = hsv_image[:, :, 2]
                        
                im_hue.append(hue.flatten())
                im_sat.append(sat.flatten())
                im_values.append(values.flatten())
                labels.append(classes.index)
    
    im_hue = np.array(im_hue)
    im_sat = np.array(im_sat)
    im_values = np.array(im_values)
    labels = np.array(labels)

    repeated_labels = np.repeat(labels, im_values.shape[1])
    im_label = np.reshape(repeated_labels, (-1, 1))
    img_hue = imagen2vector_test(im_hue)
    img_sat = imagen2vector_test(im_sat)
    img_values = imagen2vector_test(im_values)
    
    return img_hue, img_sat, img_values, im_label

folder_paths = [
    "C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"
]

img_hue, img_sat, img_values, im_label = extract_hsv_features_from_folders(folder_paths)
```

```python
import os
import numpy as np
import tifffile
from skimage import transform, color

def extract_fourier_moments_from_folders(folder_paths, target_shape=(100, 100)):
    fourier = []
    classes = ["Sanos", "Agudos", "Cronicos"]
    labels = []

    for folder_path in folder_paths:
        for class_name in classes:
            class_path = os.path.join(folder_path, class_name)
            class_index = classes.index(class_name)
            for filename in os.listdir(class_path):
                file_path = os.path.join(class_path, filename)
                image = tifffile.imread(file_path)
                resized_image = transform.resize(image, target_shape)
                
                grayscale_image = color.rgb2gray(resized_image)
                dark_fourier = np.fft.fftshift(np.fft.fft2(grayscale_image))
                                        
                fourier.append(dark_fourier.flatten())
                labels.append(class_index)

    im_fourier = np.array(fourier)
    labels = np.array(labels)

    repeated_labels = np.repeat(labels, im_fourier.shape[1])
    im_label = np.reshape(repeated_labels, (-1, 1))
    img_fourier = imagen2vector_test(im_fourier)

    return img_fourier, im_label

folder_paths = [
    "C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"
]

img_fourier, im_label = extract_fourier_moments_from_folders(folder_paths)
print('img_fourier', img_fourier.shape)
```

```python
def extract_gray_features_from_folders(folder_paths, target_shape=(100, 100)):
    im_gray = []
    classes = ["Sanos", "Agudos", "Cronicos"]
    labels = []
    
    for folder_path in folder_paths:
        for class_name in classes:
            class_path = os.path.join(folder_path, class_name)
            class_index = classes.index(class_name)
            for filename in os.listdir(class_path):
                file_path = os.path.join(class_path, filename)
                image = tifffile.imread(file_path)
                resized_image = transform.resize(image, target_shape)
                gray = rgb2gray(resized_image)

                im_gray.append(gray.flatten())
                labels.append(class_index)
    
    im_gray = np.array(im_gray)
    labels = np.array(labels)

    repeated_labels = np.repeat(labels, im_gray.shape[1])
    im_label = np.reshape(repeated_labels, (-1, 1))
    img_gray = imagen2vector_test(im_gray)
    
    return img_gray, im_label

folder_paths = [
    "C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"
]

img_gray, im_label = extract_gray_features_from_folders(folder_paths)
print('img_gray', img_gray.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, transform, util
from skimage.filters.rank import entropy
from skimage.morphology import disk

def extract_entropy_features_from_folders(folder_paths, target_shape=(100, 100)):
    im_entropy = []
    classes = ["Sanos", "Agudos", "Cronicos"]
    labels = []
    
    for folder_path in folder_paths:
        for class_name in classes:
            class_path = os.path.join(folder_path, class_name)
            class_index = classes.index(class_name)
            for filename in os.listdir(class_path):
                file_path = os.path.join(class_path, filename)
                image = tifffile.imread(file_path)
                resized_image = transform.resize(image, target_shape)
                    
                # Convertir a escala de grises
                gray_image = color.rgb2gray(resized_image)
                
                # Convertir manualmente la imagen a uint8
                gray_image_uint8 = util.img_as_ubyte(gray_image)

                # Calcular la entropía
                entropy_image = entropy(gray_image_uint8, disk(3))
                im_entropy.append(entropy_image.flatten())
                labels.append(class_index)
    
    im_entropy = np.array(im_entropy)
    labels = np.array(labels)
    repeated_labels = np.repeat(labels, im_entropy.shape[1])
    im_label = np.reshape(repeated_labels, (-1, 1))
    img_entropy = imagen2vector_test(im_entropy)
    
    return img_entropy, im_label

folder_paths = [
    "C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"
]

img_entropy, im_label = extract_entropy_features_from_folders(folder_paths)
print('img_entropy', img_entropy.shape)
```

```python
def normalize(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```python
def normalize_minmax(img_red, img_green, img_blue, img_hue, img_sat, img_values, img_fourier, img_gray, img_entropy):
    img_red_n = normalize(img_red)
    img_green_n = normalize(img_green)
    img_blue_n = normalize(img_blue)
    img_hue_n = normalize(img_hue)
    img_sat_n = normalize(img_sat)
    img_values_n = normalize(img_values)
    img_fourier_n = normalize(img_fourier)
    img_fouriern = np.real(img_fourier_n)
    img_gray_n = normalize(img_gray)
    img_entropy_n = normalize(img_entropy)
    return img_red_n, img_green_n, img_blue_n, img_hue_n, img_sat_n, img_values_n, img_fouriern, img_gray_n, img_entropy_n
```

```python
def concat(img_red_n, img_green_n, img_blue_n, img_hue_n, img_sat_n, img_values_n, img_fouriern, img_gray_n, img_entropy_n, im_label, axis=1):
    return np.concatenate([img_red_n, img_green_n, img_blue_n, img_hue_n, img_sat_n, img_values_n, img_fouriern, img_gray_n, img_entropy_n, im_label], axis=1)
```

```python
def overall(data_path, filename):
    # Aplicar las cinco funciones a cada uno de los paths especificados
    print('data_path', data_path)
    img_red, img_green, img_blue, im_label = extract_rgb_features_from_folders(data_path)
    img_hue, img_sat, img_values, im_label = extract_hsv_features_from_folders(data_path)
    img_fourier, im_label = extract_fourier_moments_from_folders(data_path)
    img_gray, im_label = extract_gray_features_from_folders(data_path)
    img_entropy, im_label = extract_entropy_features_from_folders(data_path)
    img_red_n, img_green_n, img_blue_n, img_hue_n, img_sat_n, img_values_n, img_fourier_n, img_gray_n, img_entropy_n = normalize_minmax(img_red, img_green, img_blue, img_hue, img_sat, img_values, img_fourier, img_gray, img_entropy)
    data_features = concat(img_red_n, img_green_n, img_blue_n, img_hue_n, img_sat_n, img_values_n, img_fourier_n, img_gray_n, img_entropy_n, im_label)

    columns = ["Rojo", "Verde", "Azul", "hue", "sat", "values", "fourier", "gray", "entropy", "etiquetas"]
    data_df = pd.DataFrame(data_features, columns=columns)
    
    data_df.to_csv(filename + ".csv", index=False)

# Rutas de los conjuntos de datos
train_path = ["C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Train"]
val_path = ["C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Val"]
test_path = ["C:/Users/moise/Estancia imas/data/bd_multiclass_for_ml/Test"]

# Procesar y guardar los conjuntos de datos
overall(train_path, "Train_100")
overall(val_path, "Val_100")
overall(test_path, "Test_100")
```
